```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()

for year in range(2008, 2026):  # 2026 excluded so loop goes up to 2025
    url = f"https://www.iplt20.com/matches/results/{year}"
    driver.get(url)
    time.sleep(6)  # Wait for page JS to render

    matches = driver.find_elements(By.CSS_SELECTOR, "li.ng-scope")
    results = []
    for match in matches:
        try:
            try:
                stage = match.find_element(By.CSS_SELECTOR, "span.vn-ticketTitle.ng-binding.ng-scope").text
            except:
                stage = ""
            try:
                venue = match.find_element(By.CSS_SELECTOR, "div.vn-schedule-head").text
            except:
                venue = ""
            try:
                match_datetime = match.find_element(By.CSS_SELECTOR, "span.vn-matchDateTime.ng-binding.ng-scope").text
            except:
                match_datetime = ""
            try:
                summary = match.find_element(By.CSS_SELECTOR, "div.vn-ticketTitle.ng-binding.ng-scope").text
            except:
                summary = ""

            try:
                live_score = match.find_element(By.CSS_SELECTOR, "div.live-score")
                shed_teams = live_score.find_elements(By.CSS_SELECTOR, "div.vn-shedTeam")
                team_names = []
                team_scores = []
                for shed_team in shed_teams:
                    try:
                        name = shed_team.find_element(By.CSS_SELECTOR, "div.vn-teamTitle h3.ng-binding.ng-scope").text
                    except:
                        name = ""
                    try:
                        score = shed_team.find_element(By.CSS_SELECTOR, "p.ng-binding.ng-scope").text
                    except:
                        score = ""
                    team_names.append(name)
                    team_scores.append(score)
                team1, team2 = (team_names + ["", ""])[:2]
                score1, score2 = (team_scores + ["", ""])[:2]
            except:
                team1 = team2 = score1 = score2 = ""

            results.append({
                "year": year,
                "stage": stage,
                "venue": venue,
                "date_time": match_datetime,
                "summary": summary,
                "team1": team1,
                "score1": score1,
                "team2": team2,
                "score2": score2
            })
        except Exception:
            continue

    df = pd.DataFrame(results)
    df.to_csv(f"ipl_results_{year}.csv", index=False)
    print(f"Saved results for {year}")

driver.quit()
```
