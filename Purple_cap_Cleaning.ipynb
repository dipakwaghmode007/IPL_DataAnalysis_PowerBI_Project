```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\dipak\\Innomatics\\raw_web_csv\\data\\purple_cap_all_seasons.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop(["POS"], axis=1)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df["Team"] = df["Player"].str.extract(r'([A-Z]{2,4})$')
```

```python
df["Player"] = df["Player"].str.replace(r'([A-Z]{2,4})$', '', regex=True).str.strip()
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df[["BBI_Wkts", "BBI_Runs"]] = df["BBI"].str.split("/", expand=True)
df["BBI_Wkts"] = pd.to_numeric(df["BBI_Wkts"], errors="coerce")
df["BBI_Runs"] = pd.to_numeric(df["BBI_Runs"], errors="coerce")
```

```python
df = df.drop(columns=["BBI"])
```

```python
df["Player"] = df["Player"].astype(str)
df["Team"]   = df["Team"].astype(str)
df["Wkts"] =  df["Wkts"].astype(int)
df["Runs"] =  df["Runs"].astype(int)
df["Mat"] =  df["Mat"].astype(int)
df["Inns"] =  df["Inns"].astype(int)
df["4w"] =  df["4w"].astype(int)
df["5w"] =  df["5w"].astype(int)
```

```python
df.to_csv("Cleaned_Purple_Cap_Data.csv")
```
