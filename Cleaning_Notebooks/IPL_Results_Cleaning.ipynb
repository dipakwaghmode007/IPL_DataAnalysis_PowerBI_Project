```python
df = df.drop(columns=["stage","date_time"])
```

```python
df[["match_number", "venue"]] = df["venue"].str.extract(
    r'MATCH\s+(\d+)\n([^\n]*)'
)
```

```python
df.head()
```

```python
team_shortforms = {
    r"chennai super kings": "CSK",
    r"mumbai indians": "MI",
    r"royal challengers (bangalore|bengaluru)": "RCB",
    r"rajasthan royals": "RR",
    r"punjab kings": "PBKS",
    r"delhi daredevils|delhi capitals": "DC",
    r"kolkata knight riders": "KKR",
    r"sunrisers hyderabad": "SRH",
    r"gujarat titans": "GT",
    r"lucknow super giants": "LSG",
    r"deccan chargers": "DC",
    r"pune warriors": "PW",
    r"rising pune supergiant[s]?": "RPS"
}

def replace_teams(text):
    if pd.isna(text):
        return text
    text_lower = text.lower()
    for pattern, short in team_shortforms.items():
        text_lower = re.sub(pattern, short, text_lower, flags=re.IGNORECASE)
    return text_lower.upper()

# Apply to columns
df["team1_short"] = df["team1"].apply(replace_teams)
df["team2_short"] = df["team2"].apply(replace_teams)
df["summary_short"] = df["summary"].apply(replace_teams)
```

```python
df
```

```python
# TEAM 1
df["team1_runs"] = (
    df["score1"]
    .astype(str)
    .str.extract(r"(\d+)", expand=False)
    .fillna(0)              # handle NaN safely
    .astype(int)
)

df["team1_wickets"] = (
    df["score1"]
    .astype(str)
    .str.extract(r"/(\d+)", expand=False)
    .fillna(10)             # assume all out
    .astype(int)
)

# TEAM 2
df["team2_runs"] = (
    df["score2"]
    .astype(str)
    .str.extract(r"(\d+)", expand=False)
    .fillna(0)
    .astype(int)
)

df["team2_wickets"] = (
    df["score2"]
    .astype(str)
    .str.extract(r"/(\d+)", expand=False)
    .fillna(10)
    .astype(int)
)
```

```python
df
```

```python
df = df.drop(columns=["summary", "team1", "score1", "team2", "score2"])
```

```python
df
```

```python
df["win_type"] = "Unknown"

df.loc[df["summary_short"].str.contains(r"WON BY \d+ RUN", regex=True, na=False), "win_type"] = "Defending"

df.loc[df["summary_short"].str.contains(r"WON BY \d+ WICKET", regex=True, na=False), "win_type"] = "Chasing"

df.loc[df["summary_short"].str.contains(r"SUPER OVER", regex=True, na=False), "win_type"] = "Super Over"

df.loc[df["summary_short"].str.contains(r"DLS", regex=True, na=False), "win_type"] = "DLS"

df.loc[df["summary_short"].str.contains(r"ABANDONED|ABOND", regex=True, na=False), "win_type"] = "Abandoned"
```

```python
df["match_winner"] = (
    df["summary_short"]
    .str.extract(r"^([A-Z]+)", expand=False)
)

df.loc[
    df["summary_short"].str.contains(r"ABANDONED|NO RESULT", na=False),
    "match_winner"
] = "No Result"
```
