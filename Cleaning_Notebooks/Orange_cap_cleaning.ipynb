```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\dipak\\Innomatics\\raw_web_csv\\data\\orange_cap_all_seasons.csv")
```

```python
df.head()
```

```python
df = df.drop(["Pos", "Team"], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={
    "Matches": "Runs_temp",   # temporary name to avoid conflict
    "Innings": "Matches",
    "Not Outs": "Innings",
    "Runs": "Not Outs"
})

# Now rename temporary column to final name
df = df.rename(columns={"Runs_temp": "Runs"})
```

```python
df.head()
```

```python
df["Player"].isnull().sum()
```

# Feature Extraction

```python
df["Team"] = df["Player"].str.extract(r'([A-Z]{2,4})$')
```

```python
df["Player"] = df["Player"].str.replace(r'([A-Z]{2,4})$', '', regex=True).str.strip()
```

```python
df.to_csv("Cleaned_Orange_Cap_Data.csv")
```
